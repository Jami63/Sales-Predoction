```python
import pandas as P
```

```python
D = {
    "Advertisement" : [2000 , 2500 , 3000 , 3500 , 4000] ,
    "Price" : [50 , 48 , 45 , 43 , 40] ,
    "Sales" : [500 , 550 , 620 , 700 , 820]
}
```

```python
DF = P.DataFrame(D)
```

```python
X = DF [["Advertisement" , "Price"]]
Y = DF ["Sales"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train , X_Test , Y_Train , Y_Test = train_test_split (X , Y , test_size = 0.1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
M = LinearRegression ()
```

```python
M.fit (X_Train , Y_Train)
```

```python
Prediction = M.predict (X_Test)
```

```python
print(Prediction)
```

```python
P = M.predict ([[2000 , 40]])
```

```python
print(P)
```
